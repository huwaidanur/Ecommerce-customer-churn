**global outliers**

kita akan mencari tahu konteks dari tiap global outliers, kemudian mempertimbangkannya untuk menghapusnya atau tetap membiarkannya pada dataset

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    df_no_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_no_outliers
```

Hitung skewness numerical features
* ``> 0``: Skewness positif (ekor kanan lebih panjang).
* ``< 0``: Skewness negatif (ekor kiri lebih panjang).
* ``= 0``: Simetris (normal).
* diantara ``-0.5 dan 0.5`` masih cukup simetris

Uji Normalitas menggunana normaltest
* H0 : data terdistribusi normal
* H1 : data tidak terdistribusi normal
* p-value         : 0.05 
* p-value <= 0.05 : tolak H0 

```python
cols_num = [col for col in df1.describe().columns]
```

```python
# skewness
plt.figure(figsize=(12, 8))
sns.set_palette('Greens')
for index, col in enumerate(cols_num, 1):
    plt.subplot(3,3,index)
    sns.histplot(x=df1[col], kde=True, label=f'skewness: {skew(df1[col]):.2f}', color='green')
    #sns.kdeplot(x=df[col], label=f'skewness: {skew(df[col]):.2f}', color='green')
    col_skew = skew(df1[col])
    plt.legend(loc='upper right')
    plt.title(col)
    plt.tight_layout()
plt.show()
```

```python
# normal test
pval = []
dist = []
for col in cols_num:
    stat, p_value = normaltest(df1[col])
    if p_value > 0.05:
        pval.append(p_value)
        dist.append('normal')
    else :
        pval.append(p_value)
        dist.append('non-normal')
pd.DataFrame({
    'kolom' : cols_num,
    'p_value' : pval,
    'dist' : dist
})
```

```python
# hubungan Tenure dan Churn 
# Mann Whitney U Test
from scipy.stats import mannwhitneyu

# Memisahkan data berdasarkan kategori churn
tenure_churn_0 = df1[df1['Churn'] == 0]['Tenure']
tenure_churn_1 = df1[df1['Churn'] == 1]['Tenure']

# Melakukan Mann-Whitney U test
stat, p = mannwhitneyu(tenure_churn_0, tenure_churn_1)
print(f"U-Statistic: {stat}, P-Value: {p}")

# Interpretasi hasil
if p > 0.05:
    print("Tidak ada perbedaan signifikan pada Tenure antara churn dan non-churn.")
else:
    print("Ada perbedaan signifikan pada Tenure antara churn dan non-churn.")
```

```python
def calc_error(X, y, model):
    predictions = model.predict(X)
    probas = model.predict_proba(X)[:, 1]
    return {
        'report': classification_report(y, predictions),
        'f1': f1_score(y, predictions, average='macro'),
        'roc': roc_auc_score(y, probas),
        'accuracy': accuracy_score(y, predictions),
        'recall': recall_score(y, predictions),
        'precision': precision_score(y, predictions)
    }

def calc_metrics(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    return calc_error(X_train, y_train, model), calc_error(X_test, y_test, model)

K = 10
kf = StratifiedKFold(n_splits=K, shuffle=True, random_state=42)

train_errors_without_oversampling, validation_errors_without_oversampling = [], []
train_errors_with_oversampling, validation_errors_with_oversampling = [], []
```

##### Comparing other boosting

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, recall_score, precision_score
import pandas as pd

def evaluate_model(y_true, y_pred, y_proba):
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'roc_auc': roc_auc_score(y_true, y_proba),
        'f1_score': f1_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred)
    }

def y_pred_func(model, resampling_method, X_train, y_train, X_test):
    estimator = ImbPipeline([
        ('prep', transformer),
        ('resampling', resampling_method),
        ('model', model)
    ])
    estimator.fit(X_train, y_train)
    y_pred = estimator.predict(X_test)
    y_proba = estimator.predict_proba(X_test)[:, 1]  # Probabilities for ROC AUC
    return estimator, y_pred, y_proba

resampling_methods = {
    'ros': ros,
    'rus': rus,
    'smote': smote,
    'nearmiss': nearmiss,
    'smotetomek': smote_tomek
}

results = []
models = [gboost, xgboost, lgbm]
model_names = ['Gradient Boosting Classifier', 'XGB Classifier', 'LGBM Classifier']

for method_name, resampler in resampling_methods.items():
    for model, model_name in zip(models, model_names):
        estimator, y_pred, y_proba = y_pred_func(model, resampler, X_train, y_train, X_test)
        metrics = evaluate_model(y_test, y_pred, y_proba)
        metrics['model'] = model_name
        metrics['resampling'] = method_name
        results.append(metrics)

# Convert results to a DataFrame
df_resampling = pd.DataFrame(results)

# Compute mean metrics for each model
mean_metrics = df_resampling.groupby('model') #.mean().reset_index()

# Add summary row
summary_row = mean_metrics.mean(numeric_only=True)
summary_row['model'] = 'Overall Mean'
# mean_metrics = pd.concat([mean_metrics], ignore_index=True)

df_resampling.sort_values('f1_score', ascending=False, inplace=True)
df_resampling
```
